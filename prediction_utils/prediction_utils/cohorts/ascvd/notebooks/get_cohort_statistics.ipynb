```python
import itertools
import pandas as pd
from prediction_utils.extraction_utils.database import BQDatabase
from IPython.display import display
```

```python
db = BQDatabase()
```

```python
query = """
    SELECT *
    FROM som-nero-nigam-starr.spfohl_explore.{cohort_name}
"""
```

```python
cohort_names = [
    'ascvd_starr_omop_cdm5_deid_20210130_sampled',
    'ascvd_optum_dod_2019q1_cdm_53'
]
```

### Basic statistics

```python
cohort_names = [
    'ascvd_starr_omop_cdm5_deid_20210130_sampled'
]
group_variables = ['age_group', 'race_eth', 'gender_concept_name']
outcome_variable = 'ascvd_binary'
censored_variable = 'censored_binary'
for cohort_name in cohort_names:
    df = db.read_sql_query(query.format(cohort_name=cohort_name))
    print('Cohort Name: {cohort_name}'.format(cohort_name=cohort_name))
    print('Number of patients: {}'.format(df.shape[0]))
    display(df.groupby(['has_statin_history', 'has_cvd_history']).agg(incidence=(outcome_variable, 'mean'), num_patients=(outcome_variable, 'size')))
    for group_var in group_variables:
        display(df.groupby(['has_statin_history', 'has_cvd_history', censored_variable, group_var]).agg(incidence=(outcome_variable, 'mean'), num_patients=(outcome_variable, 'size')))
    for group_vars in itertools.combinations(group_variables, 2):
        display(df.query('~ascvd_binary.isnull()').groupby(['has_statin_history', 'has_cvd_history', censored_variable] + list(group_vars)).agg(incidence=(outcome_variable, 'mean'), num_patients=(outcome_variable, 'size')))
    for group_vars in itertools.combinations(group_variables, 3):
        display(df.query('~ascvd_binary.isnull()').groupby(['has_statin_history', 'has_cvd_history', censored_variable] + list(group_vars)).agg(incidence=(outcome_variable, 'mean'), num_patients=(outcome_variable, 'size')))
```
